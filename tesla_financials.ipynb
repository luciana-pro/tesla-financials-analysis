### Analisi bilancio

Importazione dati di tesla

```python

import yfinance as yf
import pandas as pd

# Scarichiamo i dati di Tesla
tesla = yf.Ticker("TSLA")

# Otteniamo il bilancio (conto economico, stato patrimoniale, flussi di cassa)
income_statement = tesla.financials
balance_sheet = tesla.balance_sheet
cash_flow = tesla.cashflow

# Visualizziamo il conto economico
print(income_statement)
print(balance_sheet)
print(cash_flow)

# Se vuoi salvare i dati in un file Excel
with pd.ExcelWriter('tesla_financials.xlsx') as writer:
    income_statement.to_excel(writer, sheet_name='Conto Economico')
    balance_sheet.to_excel(writer, sheet_name='Stato Patrimoniale')
    cash_flow.to_excel(writer, sheet_name='Flusso di Cassa')

from google.colab import files
files.download('tesla_financials.xlsx')

# Estraiamo le attività correnti e le passività correnti dallo stato patrimoniale
```

```python
import pandas as pd

# Carichiamo il file Excel (se hai già caricato l'excel su Colab, sostituisci 'nome_file.xlsx' con il percorso corretto)
file_path = 'tesla_financials.xlsx'  # Modifica questo con il percorso del tuo file
data = pd.read_excel(file_path)

# Visualizziamo le prime righe per confermare il caricamento
data.head()
```

# Analisi di Redditività


```python
# Estrai la riga 'Gross Profit' per l'anno 2023
gross_profit_2023 = data.loc[data['Unnamed: 0'] == 'Gross Profit', pd.Timestamp('2023-12-31')].values[0]
print(f'Gross Profit 2023: {gross_profit_2023}')

total_revenue_2023 = data.loc[data['Unnamed: 0'] == 'Total Revenue', pd.Timestamp('2023-12-31')].values[0]
print(f'Total Revenue 2023: {total_revenue_2023}')

#Margine di profitto lordo
gross_margin=gross_profit_2023/total_revenue_2023
print(f'Gross Margin 2023: {gross_margin}')
```

```python
operating_income_2023 = data.loc[data['Unnamed: 0'] == 'Operating Income', pd.Timestamp('2023-12-31')].values[0]
print(f'Operating Income 2023: {operating_income_2023}')

total_revenue_2023 = data.loc[data['Unnamed: 0'] == 'Total Revenue', pd.Timestamp('2023-12-31')].values[0]
print(f'Total Revenue 2023: {total_revenue_2023}')

#Margine di profitto operativo
operating_margin=operating_income_2023/total_revenue_2023
print(f'Operating Margin 2023: {operating_margin}')
```

```python
net_income_2023 = data.loc[data['Unnamed: 0'] == 'Net Income', pd.Timestamp('2023-12-31')].values[0]
print(f'Net Income 2023: {net_income_2023}')

total_revenue_2023 = data.loc[data['Unnamed: 0'] == 'Total Revenue', pd.Timestamp('2023-12-31')].values[0]
print(f'Total Revenue 2023: {total_revenue_2023}')

#Margine di profitto netto
net_margin=net_income_2023/total_revenue_2023
print(f'Net Margin 2023: {net_margin}')
```

# Analisi della Solvibilità

```python
ebit_2023 = data.loc[data['Unnamed: 0'] == 'EBIT', pd.Timestamp('2023-12-31')].values[0]
print(f'Ebit 2023: {ebit_2023}')

interest_expense_2023 = data.loc[data['Unnamed: 0'] == 'Interest Expense', pd.Timestamp('2023-12-31')].values[0]
print(f'Interest Expense 2023: {interest_expense_2023}')

#Coperura degli interessi
interest_coverage=ebit_2023/interest_expense_2023
print(f'Interest Coverage 2023: {interest_coverage}')
```
